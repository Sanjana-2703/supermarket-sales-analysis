```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Users/sanjananeralwar/Desktop/super market analysis/SuperMarket Analysis.csv")
print(df.shape)
print(df.head())
```

```python
print("Missing values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(8,5))
city_sales = df.groupby('City')['Sales'].sum().reset_index()
sns.barplot(x='City', y='Sales', data=city_sales, palette='Blues_d')
plt.title('Total Sales by City', fontsize=16)
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
city_sales = df.groupby('City')['Sales'].sum().reset_index()
sns.barplot(x='City', y='Sales', data=city_sales, hue='City', legend=False, palette='Blues_d')
plt.title('Total Sales by City', fontsize=16)
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
product_sales = df.groupby('Product line')['Sales'].sum().reset_index()
sns.barplot(x='Sales', y='Product line', data=product_sales, hue='Product line', legend=False, palette='Greens_d')
plt.title('Sales by Product Line', fontsize=16)
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
payment = df['Payment'].value_counts()
plt.pie(payment, labels=payment.index, autopct='%1.1f%%', colors=['#66b3ff','#99ff99','#ffcc99'])
plt.title('Payment Method Distribution', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Rating'], bins=20, color='purple', kde=True)
plt.title('Customer Rating Distribution', fontsize=16)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
gender_sales = df.groupby('Gender')['Sales'].sum().reset_index()
sns.barplot(x='Gender', y='Sales', data=gender_sales, hue='Gender', legend=False, palette='coolwarm')
plt.title('Total Sales by Gender', fontsize=16)
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.strftime('%B')
month_order = ['January', 'February', 'March']

plt.figure(figsize=(8,5))
month_sales = df.groupby('Month')['Sales'].sum().reindex(month_order).reset_index()
sns.barplot(x='Month', y='Sales', data=month_sales, hue='Month', legend=False, palette='Oranges_d')
plt.title('Total Sales by Month', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```
